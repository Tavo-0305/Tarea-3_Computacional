```python
import matplotlib.pyplot as plt

# Creamos 2 vectores vacios para almacenar los datos
x_vals = []
y_vals = []
with open("rk4_results.txt", "r") as file:
    next(file)  # Se salta la primer fila pues es el encabezado del archivo
    for line in file:
        x, y = map(float, line.split())
        x_vals.append(x)
        y_vals.append(y)

# Se realiza la gr치fica de lso datos:
plt.plot(x_vals, y_vals, label="Soluci칩n RK4", color="blue", marker='o', markersize=3, linestyle='-')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Soluci칩n de la Ecuaci칩n Diferencial con RK4")
plt.legend()
plt.grid(True)
plt.show()
```
